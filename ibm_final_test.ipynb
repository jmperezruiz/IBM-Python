```python
class Tarea:
    
#Este método es el constructor de la clase Tarea. 
#Toma dos argumentos: descripcion, que es la descripción de la tarea, y completada, que es un valor booleano que indica si la tarea está completada o no.
#Si no se proporciona ningún valor para completada, se establece en False por defecto.

    def __init__(self, descripcion, completada=False): 

        self.descripcion = descripcion

        self.completada = completada

 

class GestorTareas:
#El constructor de la clase GestorTareas inicializa una lista vacía tareas.

    
    def __init__(self):

        self.tareas = []

#agregar_tarea(self, descripcion): Este método agrega una nueva tarea a la lista de tareas. 
#Toma la descripción de la tarea como argumento, crea una instancia de la clase Tarea con esa descripción y la agrega a la lista de tareas.
#Muestra un mensaje indicando que la tarea se ha agregado correctamente.

    def agregar_tarea(self, descripcion):

        tarea_nueva = Tarea(descripcion)

        self.tareas.append(tarea_nueva)

        print("Tarea agregada correctamente.")

#Este método marca una tarea específica como completada. 
#Toma la posición de la tarea en la lista como argumento, accede a esa tarea en la lista, marca la tarea como completada y muestra un mensaje indicando que la tarea se ha marcado como completada.

    def marcar_completada(self, posicion):

        try:

            tarea = self.tareas[posicion]

            tarea.completada = True

            print(f"Tarea '{tarea.descripcion}' marcada como completada.")

        except IndexError:

            print("La posición especificada no existe en la lista de tareas.")

#Este método muestra todas las tareas en la lista. 
#Si no hay tareas, muestra un mensaje indicando que no hay tareas pendientes. 
#Si hay tareas, itera sobre la lista de tareas, muestra la descripción de cada tarea y su estado (completada o pendiente).

    def mostrar_todas_las_tareas(self):

        if not self.tareas:

            print("No hay tareas pendientes.")

        else:

            for i, tarea in enumerate(self.tareas):

                estado = "Completada" if tarea.completada else "Pendiente"

                print(f"{i + 1}. {tarea.descripcion} - {estado}")

#Este método elimina una tarea específica de la lista. 
#Toma la posición de la tarea en la lista como argumento, elimina esa tarea de la lista y muestra un mensaje indicando que la tarea se ha eliminado correctamente.

    def eliminar_tarea(self, posicion):

        try:

            tarea_eliminada = self.tareas.pop(posicion)

            print(f"Tarea '{tarea_eliminada.descripcion}' eliminada correctamente.")

        except IndexError:

            print("La posición especificada no existe en la lista de tareas.")

#Esta función es la función principal que inicia el programa. Crea una instancia de la clase GestorTareas y luego entra en un bucle infinito que presenta un menú de opciones al usuario:
#Agregar nueva tarea
#Marcar tarea como completada
#Mostrar todas las tareas
#Eliminar tarea
#Salir
#Dependiendo de la opción seleccionada por el usuario, invoca los métodos apropiados del objeto GestorTareas. 
#El bucle se rompe cuando el usuario elige la opción "Salir".

def main():

    gestor = GestorTareas()

 

    while True:

        print("\n---- Gestor de Tareas ----")

        print("1. Agregar nueva tarea")

        print("2. Marcar tarea como completada")

        print("3. Mostrar todas las tareas")

        print("4. Eliminar tarea")

        print("5. Salir")

 

        opcion = input("Seleccione una opción: ")

 

        if opcion == "1":

            descripcion = input("Ingrese la descripción de la nueva tarea: ")

            gestor.agregar_tarea(descripcion)

        elif opcion == "2":

            gestor.mostrar_todas_las_tareas()

            try:

                posicion = int(input("Ingrese la posición de la tarea a marcar como completada: ")) - 1

                gestor.marcar_completada(posicion)

            except ValueError:

                print("Por favor ingrese un número válido.")

        elif opcion == "3":

            gestor.mostrar_todas_las_tareas()

        elif opcion == "4":

            gestor.mostrar_todas_las_tareas()

            try:

                posicion = int(input("Ingrese la posición de la tarea a eliminar: ")) - 1

                gestor.eliminar_tarea(posicion)

            except ValueError:

                print("Por favor ingrese un número válido.")

        elif opcion == "5":

            print("¡Hasta luego!")

            break

        else:

            print("Opción no válida. Por favor, seleccione una opción válida.")

 

if __name__ == "__main__":

    main()
    
```
